```python
dog_type = []
conf= []

def type_conf(row):
    if row['p1_dog']:
        dog_type.append(row['p1'])
        conf.append(row['p1_conf'])
    elif row['p2_dog']:
        dog_type.append(row['p2'])
        conf.append(row['p2_conf'])
    elif row['p3_dog']:
        dog_type.append(row['p3'])
        conf.append(row['p3_conf'])
    else:
        dog_type.append('Unidentifiable')
        conf.append(0)

df3_clean.apply(type_conf, axis = 1)

df3_clean['dog_type'] = dog_type
df3_clean['confidence'] = conf
```
